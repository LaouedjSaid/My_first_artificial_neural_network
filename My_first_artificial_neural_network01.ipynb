# Import libraries :

```python
import pandas as pd
from sklearn import datasets
from sklearn.model_selection import train_test_split
from keras.models import Sequential
from keras.layers import Dense
import matplotlib.pyplot as plt
from sklearn.preprocessing import LabelEncoder
from keras.utils import to_categorical
```

# We will load from sklear the iris dataset & select our features and the target :
# The iris data set consists of 50 samples from each of three species of Iris (Iris setosa, Iris virginica and Iris versicolor).
# Four features were measured from each sample: the length and the width of the sepals and petals, in centimeters. 
# Based on these four features, we build model to distinguish the species from each other.image.png

```python
# load dataset 
iris = datasets.load_iris() 
# features 
X = iris.data[:, :] 
# target 
y = iris.target 
```

```python
X
```

```python
y
```

Preprocessing :





```python
import numpy as np
from keras.utils.np_utils import to_categorical
from keras.utils import to_categorical
from tensorflow.keras.utils import to_categorical

# encode class values as integers 
encoder = LabelEncoder() 
encoder.fit(y) 
encoded_Y = encoder.transform(y) 
# convert integers to dummy variables (i.e. one hot encoded) 
#  Y= np_utils.to_categorical(encoded_Y) did not working so we use this Y=tf.keras.utils.to_categorical(encoded_Y)

import tensorflow as tf

Y=tf.keras.utils.to_categorical(encoded_Y)
```

```python
Y
```

Split :

```python
X_train, X_test, y_train,y_test = train_test_split(X, Y, test_size = 0.1)
X_model, X_valid, y_model, y_valid = train_test_split(X_train, y_train, test_size = 0.2)
```

This is an Artifitial neural network where :

The First hidden layer has 16 neurons
Having 4 features (the length and the width of the sepals and petals, in centimeters) the input dimension is 4
The output layer has 3 neurons since we have 3 species of flowers

```python
# Neural network
model= Sequential([
   Dense(16, input_dim = 4, activation='relu'),    
   Dense(3, activation='softmax') ])
```

Compile & fit :
Let's now configure our Keras model for training

```python
model.compile(loss='categorical_crossentropy', optimizer='adam', metrics=["accuracy"]) # Configure the model for training
history = model.fit(X_model, y_model, validation_data=(X_valid, y_valid), epochs= 150) #training
```

Summary :
Once a model is built, you can call its summary() method to display its contents including type of layers and parameters (weights+bias to be defined after training).

```python
model.summary()
```

The history object is returned from calls to the fit() function used to train the model. Model history records accuracy and loss metrics in each epoch for both validation and training dataset.

```python
# summarize history for accuracy
plt.plot(history.history['accuracy'])
plt.plot(history.history['val_accuracy'])
plt.title('model accuracy')
plt.ylabel('accuracy')
plt.xlabel('epoch')
plt.legend(['train', 'test'], loc='upper left')
plt.show()

# summarize history for loss
plt.plot(history.history['loss'])
plt.plot(history.history['val_loss'])
plt.title('model loss')
plt.ylabel('loss')
plt.xlabel('epoch')
plt.legend(['train', 'test'], loc='upper left')
plt.show()
```
